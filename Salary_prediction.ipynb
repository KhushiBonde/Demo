```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

x1=df[['Experience Years']]
y1=df['Salary']
plt.figure(figsize=(10,3))
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.scatter(x1, y1)
plt.show()
```

```python
df.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,train_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
Year=int(input("Enter Your Experience in Year:"))
Salary=lr.predict([[Year]])
print(Salary)
```
